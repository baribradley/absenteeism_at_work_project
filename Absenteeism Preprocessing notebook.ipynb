# Preprocessing Employee Data

```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('Absenteeism_data.csv')
raw_data.head()
```

```python
raw_data.shape
```

```python
data = raw_data.copy()
```

```python
pre_data = data.drop(['ID'], axis=1)
pre_data.head()
```

### Reason of Absence
#### Each number of the 'Reason for Absence' column represents 1 specific reason why an employee was absent. I will create dummny variables for quantitative analysis purposes and group these features based on their similarities.  Reason_1 = Reasons based on diseases, Reason_2 = related to pregnancy or giving birth, Reason_3 = Reasons having to do with poisoning, Reason_4 = Light reasons for absence(physical therapy, medical consultations etc.)

```python
sorted(pre_data['Reason for Absence'].unique())
```

```python

dummy_var_1 = pd.get_dummies(pre_data['Reason for Absence'], drop_first=True)
dummy_var_1
```

```python
dummy_var_1[dummy_var_1.columns[:14]]
```

```python
reason_type_1 = dummy_var_1.loc[:,1:14].max(axis=1)
reason_type_2 = dummy_var_1.loc[:,15:17].max(axis=1)
reason_type_3 = dummy_var_1.loc[:,18:21].max(axis=1)
reason_type_4 = dummy_var_1.loc[:,22:].max(axis=1)
```

```python
reason_type_1
```

```python
pre_data.columns.values
```

```python
#Combining all of the reason dataframes to the main dataframe
pre_data = pd.concat([pre_data, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
pre_data.head()
```

```python
pre_data.columns.values
```

```python
#Renaming columns
column_names = ['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
pre_data.columns = column_names
pre_data.head()
```

```python
columns_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
```

```python
pre_data = pre_data[columns_reordered]
pre_data = pre_data.drop(['Reason for Absence'], axis=1)
pre_data.head()
```

```python
#Creating a checkpoint
df_reason_mod = pre_data.copy()
```

### Date
#### First I will change the object type from string to timestamp, then extract the Month and Day of the week value for analysis purposes

```python
df_reason_mod['Date']
```

```python
#Changing dates in the "Date" column from string to timestamp object
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
df_reason_mod['Date'].head()
```

```python
type(df_reason_mod['Date'][0])
```

```python
df_reason_mod.info()
```

```python
df_reason_mod['Date'][0].month
```

```python
#Creating a for loop to extract the month
list_months = []
for i in range(700):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df_reason_mod['Month Value'] = list_months
df_reason_mod.head()
```

```python
#Creating a simply function to extract the day of the week and create a new column
def day_of_week(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(day_of_week)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod.columns.values
```

```python
new_column_names = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Month Value',
       'Day of the Week', 'Date','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[new_column_names]
df_reason_mod = df_reason_mod.drop(['Date'], axis=1)
df_reason_mod
```

### Education

#### 1 = High School, 2 = Graduate, 3 = Post Graduate, 4 = A master of Doctor. Since most of the oberservations have only a high school level education, we can split these oberservations into two groups(0 = High School, 1 = All other degrees)

```python
df_reason_mod['Education'].value_counts()
```

```python
df_reason_mod['Education'] = df_reason_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
df_reason_mod['Education'].value_counts()
```

```python
df_preprocessed = df_reason_mod.copy()
```
